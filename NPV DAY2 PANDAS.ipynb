```python
df2=df1.sort_index(axis=0,ascending=True,inplace=True)#inplace is True-change the orginal df1 so not to be used
print(df2)
print()
print(df1)
```

```python
df2=df1.sort_index(axis=0,ascending=False,inplace=False)#inplace is 
print(df2)
print()
print(df1)
```

```python
df2=df1.sort_index(axis=1,ascending=True,inplace=False)#inplace is always False
print(df2)
print()
print(df1)
```
